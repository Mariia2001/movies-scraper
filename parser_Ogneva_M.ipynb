```python
from scrapy.crawler import CrawlerProcess
from scrapy.spiders import Spider
import csv
import os

class MoviesSpider(Spider):
    name = 'movies'
    start_urls = ['https://ru.wikipedia.org/wiki/%D0%9A%D0%B0%D1%82%D0%B5%D0%B3%D0%BE%D1%80%D0%B8%D1%8F:%D0%A4%D0%B8%D0%BB%D1%8C%D0%BC%D1%8B_%D0%BF%D0%BE_%D0%B0%D0%BB%D1%84%D0%B0%D0%B2%D0%B8%D1%82%D1%83']

    def parse(self, response):
        for movie_link in response.css('div.mw-category-group ul li a::attr(href)').extract():
            yield response.follow(movie_link, callback=self.parse_movie)

    def parse_movie(self, response):
        title = response.css('h1.firstHeading::text').get()
        genre = response.css('td:contains("Жанр") + td a::text').extract_first()
        director = response.css('td:contains("Режиссёр") + td a::text').extract_first()
        country = response.css('td:contains("Страна") + td a::text').extract_first()
        year = response.css('td:contains("Год") + td a::text').extract_first()

        # Выводим результаты на экран
        print(f'Title: {title}, Genre: {genre}, Director: {director}, Country: {country}, Year: {year}')

        # Сохраняем в CSV файл
        self.save_to_csv(title, genre, director, country, year)

    def save_to_csv(self, title, genre, director, country, year):
        file_path = 'movies.csv'
        fieldnames = ['Title', 'Genre', 'Director', 'Country', 'Year']

        # Если файл не существует, записываем заголовки
        if not os.path.exists(file_path):
            with open(file_path, 'w', newline='', encoding='utf-8') as csvfile:
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                writer.writeheader()

        # Записываем данные о фильме в CSV файл
        with open(file_path, 'a', newline='', encoding='utf-8') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writerow({'Title': title, 'Genre': genre, 'Director': director, 'Country': country, 'Year': year})

# Запускаем парсер
process = CrawlerProcess(settings={
    'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
})
process.crawl(MoviesSpider)
process.start()
```
